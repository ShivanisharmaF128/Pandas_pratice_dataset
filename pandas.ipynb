```python
data ={"rollno":[101,102,103,104],
       "name":['vani','rakhi','raj','tanya'],
       "class":[1,2,3,4],
       "salary":['55k','20k','40k','30k']}
df2=pd.DataFrame(data)
df2
```

```python
Data1 ={"rollno":[101,102,103,105],
       "name":['vani','rakhi','raj','tanya'],
       "grade":['A','B','C','D'],
       "salary":['55k','20k','40k','30k']}
df3=pd.DataFrame(Data1)
df3
```

```python
df4=df2.merge(df3,on='rollno',how='inner') #Matching values of both the tables.
df4
```

```python
df4=df2.merge(df3,on='rollno',how='left') #Matching values of both the tables and all the values of left table
df4
```

```python
df4=df2.merge(df3,on='rollno',how='right') #Matching values of both the tables and all the values of right table
df4
```

```python
df4=df2.merge(df3,how='cross') 
df4   #how='cross' means: “ignore any common columns, take every row from the first DataFrame and combine it with every row of the second DataFrame” (a Cartesian product).
```
